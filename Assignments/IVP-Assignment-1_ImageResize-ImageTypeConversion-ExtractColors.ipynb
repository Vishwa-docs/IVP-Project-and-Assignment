### Image Resizing

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open("input_img.jpeg")

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
```

```python
def nearest_neighbor(inp, op, new_size):
    og_image = Image.open(inp)
    og_pixels = np.array(og_image)

    new_h, new_w = new_size
    new_pixels = np.zeros((new_h, new_w, og_pixels.shape[2]), dtype=np.uint8)

    for i in range(new_h):
        for j in range(new_w):
            og_i = int(i * og_pixels.shape[0] / new_h)
            og_j = int(j * og_pixels.shape[1] / new_w)
            new_pixels[i, j] = og_pixels[og_i, og_j]

    resized_image = Image.fromarray(new_pixels)
    
    plt.subplot(1, 2, 2)
    plt.imshow(resized_image)
    plt.title("Resized Image")
    plt.axis('off')
    plt.show()
```

```python
nearest_neighbor('input_img.jpeg', 'output_img.jpeg', (300, 200))
```

```python
nearest_neighbor('input_img.jpeg', 'output_img.jpeg', (40,40))
```

```python
nearest_neighbor('input_img.jpeg', 'output_img.jpeg', (500,100))
```

```python
nearest_neighbor('input_img.jpeg', 'output_img.jpeg', (100,500))
```

### Image Type Conversions

```python
def convert_format(inp, op, op_format):
    og_image = Image.open(inp)
    og_image.save(op, format=op_format)
```

```python
convert_format('input_img.jpeg', 'output.png', 'PNG')
```

```python
convert_format('input_img.jpeg', 'output.gif', 'GIF')
```

### Color Band

```python
def extract_color_bands(inp):
    og_image  = Image.open(inp)
    og_pixels = np.array(og_image)

    blue_channel  = og_pixels[:, :, 0]
    green_channel = og_pixels[:, :, 1]
    red_channel   = og_pixels[:, :, 2]

    return blue_channel, green_channel, red_channel

# Example usage:
blue, green, red = extract_color_bands('input_img.jpeg')
print("Blue : \n",blue)
print()
print("Green : \n",green)
print()
print("Red : \n",red)
```
